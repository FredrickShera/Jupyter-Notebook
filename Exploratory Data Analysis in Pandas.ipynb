```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\hp\OneDrive\Documents\Pandas Tutorial Files\world_population.csv')
df
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df.info() #provides high value information
```

```python
df.describe() #gets a hig overview of the data
```

```python
df.isnull().sum() #checks missing values
```

```python
df.nunique() #shows unique values in our data set
```

```python
df.sort_values(by = 'World Population Percentage', ascending = False).head(10)
```

```python
df.corr(numeric_only = True)
```

```python
sns.heatmap(df.corr(numeric_only = True), annot = True)

plt.rcParams['figure.figsize'] = (20,7)
plt.show()
```
