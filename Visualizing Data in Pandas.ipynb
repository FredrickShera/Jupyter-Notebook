# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv(r'C:\Users\hp\OneDrive\Documents\Pandas Tutorial Files\Ice Cream Ratings.csv')
df.set_index('Date')
df
```

```python
print(plt.style.available)
```

```python
df.plot(kind = 'line', subplots = False)
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y= 'Overall Rating', s = 300, c = 'red')
```

```python
df.plot.hist(bins = 20)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10,10))
```
